# Resultados del Experimento de Modelado

Este notebook resume y compara los resultados obtenidos en los diferentes experimentos de predicción de unidades vendidas.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Centralizar métricas de cada modelo probado
results = pd.DataFrame([
    {"Modelo": "Random Forest", "MAE": 3.04, "RMSE": 4.22},
    {"Modelo": "LightGBM", "MAE": 2.84, "RMSE": 4.22},
    {"Modelo": "XGBoost", "MAE": 2.93, "RMSE": 4.29},
    {"Modelo": "LSTM", "MAE": 3.31, "RMSE": 4.63},
    {"Modelo": "Exponential Smoothing", "MAE": None, "RMSE": None}
])

results.style.format({"MAE": "{:.2f}", "RMSE": "{:.2f}"}).highlight_min(
    subset=["MAE", "RMSE"], color="grey"
)

```

```python
# Comparar MAE
plt.figure(figsize=(8,5))
plt.bar(results["Modelo"], results["MAE"], color="skyblue")
plt.title("Comparación de MAE por Modelo")
plt.ylabel("MAE (menor es mejor)")
plt.xticks(rotation=30)
plt.show()

# Comparar RMSE
plt.figure(figsize=(8,5))
plt.bar(results["Modelo"], results["RMSE"], color="salmon")
plt.title("Comparación de RMSE por Modelo")
plt.ylabel("RMSE (menor es mejor)")
plt.xticks(rotation=30)
plt.show()
```

```python
# Ordenar resultados por desempeño
results_sorted = results.dropna().sort_values("MAE")
results_sorted
```

## Conclusión General

- **LightGBM** fue el modelo con mejor desempeño global (MAE = 2.84, RMSE = 4.22), mostrando un buen equilibrio entre precisión y eficiencia.  
- **XGBoost** fue competitivo y sólido, con métricas cercanas a LightGBM, aunque con un RMSE ligeramente mayor.  
- **Random Forest** obtuvo resultados aceptables, capturando patrones, aunque menos eficiente frente a los métodos de boosting.  
- **LSTM** presentó resultados inestables y menor desempeño en este conjunto de datos, lo que sugiere necesidad de mayor ajuste o más datos.  
- **Exponential Smoothing** no resultó adecuado para datos granulares, mostrando limitaciones al generalizar entre productos.  

**Modelo sugerido para continuar: LightGBM.**
